```python
texte = input("Entrez un mot ou une phrase : ")

print(f"Longueur : {len(texte)} caractères")
print(f"En majuscules : {texte.upper()}")
print(f"En minuscules : {texte.lower()}")
print(f"Inversé : {texte[::-1]}")

nettoye = texte.replace(" ", "").lower()
inverse = nettoye[::-1]

if nettoye == inverse:
    print("C'est un palindrome.")
else:
    print("Ce n'est pas un palindrome.")


print("Les méthodes comme upper(), lower() ne modifient PAS la chaîne originale, elles créent une nouvelle chaîne, car les strings sont immuables en Python.")
```
